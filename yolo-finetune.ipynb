```python
# https://docs.ultralytics.com/tasks/detect/#how-do-i-train-a-yolo26-model-on-my-custom-dataset

from ultralytics import YOLO
import torch
import numpy as np


# Load a pretrained model
model = YOLO("/kaggle/input/models/kevinohmru/10epoch/pytorch/default/1/best.pt")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
# Train the model on your custom dataset
# https://github.com/ultralytics/ultralytics/blob/main/ultralytics/cfg/datasets/hand-keypoints.yaml
model.train(
    data="/kaggle/input/datasets/kevinohmru/hand-keypoints/hand-keypoints.yaml",
    epochs=20,
    imgsz=640,
    task="pose",
    device=0
)

```
